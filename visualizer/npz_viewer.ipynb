```python
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

SCENARIO_PATH = Path("Scenic/examples/gta/twoCars.scenic")
PNG_PATH = SCENARIO_PATH.with_name("map.png")

img = mpimg.imread(PNG_PATH)   # (H, W, 3 or 4)

H, W = img.shape[:2]
print("size:", W, H)

fig = plt.figure(figsize=(6, 10), facecolor="black")
ax = plt.gca()
ax.set_facecolor("black")

ax.imshow(img)

ax.set_xlabel("x (pixels)")
ax.set_ylabel("y (pixels)")

ax.set_ylim(H, 0)

step_x = 200
step_y = 200
ax.set_xticks(range(0, W+1, step_x))
ax.set_yticks(range(0, H+1, step_y))

ax.tick_params(colors="white")
ax.xaxis.label.set_color("white")
ax.yaxis.label.set_color("white")
for spine in ax.spines.values():
    spine.set_color("white")

ax.grid(False)

plt.tight_layout()
plt.savefig("map_coordinates.png", dpi=150, facecolor=fig.get_facecolor(), edgecolor="black")
plt.show()
```
